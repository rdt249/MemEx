```python
from memex import *

def check_sel(i1,i2,i3,i4,threshold=1000) :
    status = ''
    if abs(i1)>threshold :
        set_out1(0) 
        status += 'LATCHUP DETECTED - CHANNEL A\n'
    if abs(i2)>threshold :
        set_out2(0)
        status += 'LATCHUP DETECTED - CHANNEL B\n'
    if abs(i3)>threshold :
        set_out3(0)
        status += 'LATCHUP DETECTED - CHANNEL C\n'
    if abs(i4)>threshold :
        set_out4(0)
        status += 'LATCHUP DETECTED - CHANNEL D\n'
    return status

def auto_seu(inc=60,pattern=85,vhold=0.6,vsweep=[None,None,None],slimit=None) :
    size = 32 # size in kB
    active,awake = rollcall()
    pins = [board.D5,board.D6,board.D13,board.D19,board.D26,board.D24,
            board.D25,board.D7,board.D12,board.D16,board.D20,board.D21]
    pins = [x for (x, y) in zip(pins, awake) if y]
    slots = ['A1','A2','A3','B1','B2','B3','C1','C2','C3','D1','D2','D3']
    slots = [x for (x, y) in zip(slots, awake) if y]
    #print(pins,slots)
    addr = []
    flips = []
    meta = []
    print()
    if all(vsweep) :
        vrange = list(np.arange(vsweep[1],vsweep[0]+vsweep[2],vsweep[2])[::-1])
        vrange = [round(x,3) for x in vrange]
        print('vrange =',vrange)
    v = vhold
    print('V','S',*slots,sep='\t')
    trial = 0
    sweep = 0
    while True :
        set_outs(3.3,3.3,3.3,1.8)
        if any(vsweep) :
            v = vrange[trial]
        comms_on()
        m = [0] * len(slots)
        for i in range(len(slots)) :
            m[i] = sram(cs=pins[i],name=slots[i],size=size)
            m[i].fill(pattern)
        comms_off()
        set_outs(v,v,v,v)
        #read_outs()
        start = time.time()
        while time.time() - start < inc :
            ### latchup detection here
            pass
        set_outs(3.3,3.3,3.3,1.8)
        #read_outs()
        comms_on()
        data = pd.DataFrame(columns=slots)
        bits = [0] * len(slots)
        check = bytes2bits([pattern] * (1024 * size))
        for i in range(len(slots)) :
            data[slots[i]] = bytes2bits(m[i].export())
            data[slots[i]] = [x!=y for x,y in zip(data[slots[i]],check)]
            bits[i] = list(data[slots[i]].where(data[slots[i]]==True).dropna().index)
        addr += [bits]
        pd.DataFrame(addr,columns=slots).to_csv('data/sram/seu/addr.csv',index=False)
        print(v,sweep,*data.sum().values,sep='\t')
        flips += [list(data.sum())]
        pd.DataFrame(flips,columns=slots).to_csv('data/sram/seu/flips.csv',index=False)
        meta += [[int(time.time()),inc,pattern,v,sweep,data.sum().sum(),size]]
        pd.DataFrame(meta,columns=['Timestamp(s)','Increment(s)','Pattern(byte)','Bias(v)','Sweep(#)','SEUs(#)',
                                   'Size(kB)']).to_csv('data/sram/seu/meta.csv',index=False)
        if trial == len(vrange)-1 :
            trial = 0
            sweep += 1
            if slimit is not None :
                if sweep == slimit :
                    break
        else :
            trial += 1
            
def auto_sweep(vsweep=[3.3,0.6,0.3],seu_limit=1000,inc=60,pattern=85,folder='data/sram/seu/auto/') :
    size = 32 # size in kB
    active,awake = rollcall()
    pins = [board.D5,board.D6,board.D13,board.D19,board.D26,board.D24,
            board.D25,board.D7,board.D12,board.D16,board.D20,board.D21]
    pins = [x for (x, y) in zip(pins, awake) if y]
    slots = ['A1','A2','A3','B1','B2','B3','C1','C2','C3','D1','D2','D3']
    slots = [x for (x, y) in zip(slots, awake) if y]
    #print(pins,slots)
    addr = []
    flips = []
    meta = []
    print()
    vrange = list(np.arange(vsweep[1],vsweep[0]+vsweep[2],vsweep[2])[::-1])
    vrange = [round(x,3) for x in vrange]
    print('vrange =',vrange)
    print()
    print('V',*slots,sep='\t')
    for v in vrange :
        seu_total = [0] * len(slots)
        while any([x < seu_limit for x in seu_total]) : # while any seu_total < seu_limit
            set_outs(3.3,3.3,3.3,1.8)
            comms_on()
            m = [0] * len(slots)
            for i in range(len(slots)) :
                m[i] = sram(cs=pins[i],name=slots[i],size=size)
                m[i].fill(pattern)
            comms_off()
            set_outs(v,v,v,v)
            #read_outs()
            start = time.time()
            while time.time() - start < inc :
                ### latchup detection here
                pass
            set_outs(3.3,3.3,3.3,1.8)
            #read_outs()
            comms_on()
            data = pd.DataFrame(columns=slots)
            bits = [0] * len(slots)
            check = bytes2bits([pattern] * (1024 * size))
            for i in range(len(slots)) :
                data[slots[i]] = bytes2bits(m[i].export())
                data[slots[i]] = [x!=y for x,y in zip(data[slots[i]],check)]
                bits[i] = list(data[slots[i]].where(data[slots[i]]==True).dropna().index)
                seu_total[i] += data[slots[i]].sum()
            addr += [bits]
            pd.DataFrame(addr,columns=slots).to_csv(folder + 'addr.csv',index=False)
            print(v,*data.sum().values,sep='\t')
            flips += [list(data.sum())]
            pd.DataFrame(flips,columns=slots).to_csv(folder + 'flips.csv',index=False)
            meta += [[int(time.time()),inc,pattern,v,data.sum().sum(),size]]
            pd.DataFrame(meta,columns=['Timestamp(s)','Increment(s)','Pattern(byte)','Bias(v)','SEUs(#)',
                                       'Size(kB)']).to_csv(folder + 'meta.csv',index=False)
    print()
    print('end of sweep. output saved in',folder)
    print()
        
auto_sweep(vsweep=[0.5,0.48,0.01],seu_limit=100,inc=1)
```
